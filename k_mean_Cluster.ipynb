```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import metrics
```

```python
data=pd.read_csv('k-mean-dataset-mall-customer.csv')
```

```python
data.head()
```

```python
data.drop('Genre',axis=1,inplace=True)
```

```python
data
```

```python
data.drop('Age',axis=1,inplace=True)
```

```python
data
```

```python
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], s = 50, c = 'b')
plt.show()
```

```python
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(data)
    distortions.append(kmeanModel.inertia_)
 
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.annotate('Elbow',xy=(5,50000),xytext=(5,80000),arrowprops=dict(facecolor='blue',shrink=0.05))
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
Kmean = KMeans(n_clusters=5,init='k-means++',random_state=42)
```

```python
Kmean.fit(data)
```

```python
Kmean.cluster_centers_
```

```python
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], s = 50, c =Kmean.labels_)
plt.scatter(Kmean.cluster_centers_[:,0],Kmean.cluster_centers_[:,1],c='green',s=500,marker='*',edgecolors='black')
# plt.scatter(55.2962963, 49.51851852, s=200, c='g', marker='s')
# plt.scatter(88.2, 17.11428571, s=200, c='r', marker='s')
# plt.scatter(26.30434783, 20.91304348, s=200, c='y', marker='s')
# plt.scatter(25.72727273, 79.36363636, s=200, c='black', marker='s')
# plt.scatter(85.21052632, 82.10526316, s=200, c='black', marker='s')
plt.show()
```
